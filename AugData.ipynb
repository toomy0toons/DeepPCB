```python
from IPython.display import Image
import matplotlib.pyplot as plt
import os
import numpy as np
import cv2
import matplotlib.image as mpimg
import traceback

basedir = "PCBData/"
d = "test.txt"

target = open(os.path.join(basedir + d), 'r') 
files = target.readlines()

annot = []
temp_imgs = []
test_imgs = []

#load file list 
for file in files:
    img,txt = file.strip().split(' ')
    temp_imgs.append(basedir+img[:-4]+'_temp.jpg')
    test_imgs.append(basedir+img[:-4]+'_test.jpg')
    annot.append(txt)
# open file

class_key = {1:'open',2:'short',3:'mousebite',4:'spur',5:'copper',6:'pin_hole'}
savepath = 'AugData/Test/test0217'

if not os.path.exists(savepath):
    os.makedirs(savepath)
    print(savepath + ' folder created')
for val in class_key.values():
    if not os.path.exists(os.path.join(savepath,val)):
        os.makedirs(os.path.join(savepath,val)) 
        print(val + ' folder created')
        
        
count = 0
for temp,test,txt in zip(temp_imgs,test_imgs,annot):

    not_file = open(os.path.join(basedir+txt),'r')
    notations = not_file.readlines()
    img1 = mpimg.imread(temp)
    img2 = mpimg.imread(test)
    for box in notations:
        x1,y1,x2,y2,defect = box.strip().split(' ')
        
        try:
            replace = img1.copy()
            replace[int(y1):int(y2),int(x1):int(x2)] = img2[int(y1):int(y2),int(x1):int(x2)]
            rgb = cv2.cvtColor(replace, cv2.COLOR_BGR2RGB)
            name = temp[-18:-9] + '_'+ x1 + '_'+ y1 + '_'+ x2 + '_'+ y2 + '_'+ defect + '.jpg'
            mpimg.imsave(savepath+'/'+class_key[int(defect)]+name,rgb)
        except Exception:
            traceback.print_exc()
            print(x1,y1,x2,y2,defect)
            print(temp)
            plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
            plt.show()
            print(test)
            plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
            plt.show()
            break
        if count % 100 == 0:
            print(count,name)
        count+=1
print("total count")
print (count)
        
```
